```python
from econometrics_stats import load_data, preprocess_ln, linear_regression, instrument_ols, panel_ols
```

```python
# 데이터 로드
df_original = load_data('/content/Mock_Exam_2021_Data_Guns (1).xlsx')

# ln(vio)가 있으니까 preprocess_ln 를 해줘야댐

# 일단 df_1 로 지정
# vio 만 ln 하면 됨
df_1 = preprocess_ln(['vio'], df_original)

# regression - (1) shall
linear_regression(['shall'], 'ln_vio', df_1)

# regression - (2) shall, incarc_rate, density, avginc, pop, pb1064, pw1064 and pm1029
linear_regression(['shall', 'incarc_rate', 'density', 'avginc', 'pop', 'pb1064', 'pw1064', 'pm1029'], 'ln_vio', df_1)
```

```python
# regression (3) - Do the results change when you add fixed state effects? If so, which set of regression results is more credible, and why?
# regression (4) - Do the results change when you add fixed time effects? If so, which set of regression results is more credible, and why?

# 하나씩
panel_ols(['shall', 'incarc_rate', 'density', 'avginc', 'pop', 'pb1064', 'pw1064', 'pm1029'], 'ln_vio', df_1, fixed_entity='stateid')
panel_ols(['shall', 'incarc_rate', 'density', 'avginc', 'pop', 'pb1064', 'pw1064', 'pm1029'], 'ln_vio', df_1, fixed_time='year')

# 두개
panel_ols(['shall', 'incarc_rate', 'density', 'avginc', 'pop', 'pb1064', 'pw1064', 'pm1029'], 'ln_vio', df_1, 'stateid', 'year')
```
